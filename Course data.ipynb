```python
from selenium import webdriver
import time

driver = webdriver.Chrome()
driver.get('https://www.guvi.in/courses/?current_tab=paidcourse');
time.sleep(5)

course=driver.find_elements(By.XPATH,"//div[@class='d-flex cardTitle']")

duration=driver.find_elements(By.CSS_SELECTOR,"p[class='d-inline mr-2']")
```

```python
for i in duration:
    print(i.text)
```

```python
course_data=[]
course_data.append(course)
course_data.append(duration)
```

```python
for x in course_data:
    print(x)
```
